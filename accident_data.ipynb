```python
import pandas as pd
import numpy as np

data_accident=pd.read_csv("accidents_2005_to_2007.csv")

#converting column count in to a dictionary
dic_col_count=dict(data_accident.count())
dic_col_count
```

```python
#Getting keys and values separately in to a list
list_values=list(dic_col_count.values())
list_keys=list(dic_col_count.keys())

#Getting mean column count of the columns
mean_count=np.mean(data_accident.count())
to_remove_col=[]
for i in range(len(list_values)):
    if list_values[i]< mean_count:
        to_remove_col.append(list_keys[i])

#Printing columns that have to be removed
print(to_remove_col)
```

```python
#Dropping unnecessary columns
data = data_accident.drop(to_remove_col,axis=1)
```

```python
#removing null values
data=data.dropna()
data.head()
```

```python
#converting data frame in to a csv file
data.to_csv("cleaned_data.csv")
```
